```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from deepnote import MusicRepr, Constants, Metric, Note, utils
from deepnote.metrics import pitch_histogram_entropy
```

```python
const = Constants(
    unit=4,
    min_tempo=20,
    max_tempo=200,
    num_tempo_bins=20, 
    num_velocity_bins=20
)
seq = MusicRepr.from_file('assets/test.mid', const=const)
```

```python
ent, probs = pitch_histogram_entropy(seq, window=10, pitch_class=True)


plt.figure(figsize=(20,8))
plt.subplot(2,1,1)
plt.title('entropy')
plt.plot(ent)
plt.subplot(2,1,2)
plt.title('probs')
plt.imshow(probs)
plt.show()
```
